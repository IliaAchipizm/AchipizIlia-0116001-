```python
#%% EJERCICIO 6
#Leer un número entero y determinar si la suma de sus dígitos es un número
# de Fibonacci
lista=[]
def fib(n):
    a,b = 0,1 
    while a<n:
        lista.append(a)
        a,b= b, a+b 
    return lista
num=input('Ingrese un número:')
suma=0
for i in num:
    suma=suma+int(i)
m=1000
fibo=fib(m)
bandera=0
for i in fibo:
    if i ==suma:
        bandera=1
if bandera ==1:
    print('La suma del número se encuentra en la sucesión de fibonacci ')
else:
    print('La suma del número no se encuentra en la sucesión de fibonacci ')
    
```

```python
#%% 12. Hacer un programa que lea dos palabras (podrían estar en mayúscula
# o minúscula) y determine cuál está primero en el diccionario.
palabra1=(input("Favor ingrese la palabra 1: ")).lower()
palabra2=(input("Favor ingrese la palabra 2: ")).lower()

palabra1_cop=""
palabra2_cop=""

pal_mayor= ""
espacios = 0

if len(palabra1) > len(palabra2):
    len_mayor=len(palabra1)
    espacios = len(palabra1) - len(palabra2)
    palabra2 += " " * espacios
else:
    len_mayor = len(palabra2)
    espacios = len(palabra2) - len(palabra1)
    palabra1+= " " * espacios
    
for i in palabra1:
    if 97<= ord(i) <=122: 
        palabra1_cop += i
    elif i in "á":
        palabra1_cop += "a"
    elif i in "é":
        palabra1_cop += "e"
    elif i in "í":
        palabra1_cop += "i"
    elif i in "ó":
        palabra1_cop += "o"
    elif i in "ú":
        palabra1_cop += "u"
    elif i in "ñ":
        palabra1_cop += "ñ"
    else:
        palabra1_cop += i
        
for i in palabra2: 
    
    if 97<= ord(i) <=122:
        palabra2_cop += i
    elif i in "á":
        palabra2_cop += "a"
    elif i in "é":
        palabra2_cop += "e"
    elif i in "í":
        palabra2_cop += "i"
    elif i in "ó":
        palabra2_cop += "o"
    elif i in "ú":
        palabra2_cop += "u"
    elif i in "ñ":
        palabra2_cop += "ñ"
    else:
        palabra2_cop += i

for i in range(len_mayor):
    if (palabra1_cop[i] == "ñ") and (ord(palabra2_cop[i])<= ord("n")):
        pal_mayor=palabra2
        break
    elif (palabra2_cop[i] == "ñ") and (ord(palabra1_cop[i])<= ord("n")):
        pal_mayor=palabra1
        break
    elif palabra1_cop[i] < palabra2_cop[i]:
        pal_mayor=palabra1
        break
    elif palabra1_cop[i] > palabra2_cop[i]:
        pal_mayor=palabra1
        break
if pal_mayor:
    print("la  primera palabra en el diccionario es:",{pal_mayor})   
else:
    print("las palabras son las iguales")
```

```python
#%% #  13 Escribir un programa que lea una cadena de texto y la imprima como un triángulo
cadena = input('escriba la cadena de texto: ')

cadena = cadena.replace(' ', '')


numero_caract = len(cadena)

#  núm de filas y el núm máx de caracteres por fila
t = 0                       # Variable auxiliar 
n_filas = 1                  
while t < numero_caract:
    t += 2 * n_filas - 1
    n_filas += 1

num_max = 2 * n_filas - 1    # N máx de caracteres en una fila

# Se imprime la cadena de texto 
auxi_1 = 0        
auxi_2 = 0        
contador = 0     

for i in range(1,n_filas):
    auxi_2 += 2*i-1     
    print(cadena[auxi_1:auxi_2].center(num_max,' '))
    auxi_1 = auxi_2      
    contador += 1
    
if contador != numero_caract:
    print(cadena[auxi_2:].center(num_max,' '))
```
